```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Create Sample Dataset
locations = ["Chennai", "Bangalore", "Hyderabad", "Coimbatore", "Kochi"]
items = ["Pizza", "Burger", "Biryani", "Dosa", "Fried Rice", "Pasta", "Shawarma"]

data = []

for i in range(1, 121):
    hour = random.randint(8, 22)
    minute = random.randint(0, 59)

    item = random.choice(items)

    if item == "Pizza":
        amount = 250
    elif item == "Burger":
        amount = 180
    elif item == "Biryani":
        amount = 300
    elif item == "Dosa":
        amount = 120
    elif item == "Fried Rice":
        amount = 220
    elif item == "Pasta":
        amount = 280
    else:
        amount = 160

    data.append([
        i,
        f"{hour:02d}:{minute:02d}",
        random.choice(locations),
        item,
        amount
    ])

df = pd.DataFrame(
    data,
    columns=["Order_ID", "Order_Time", "Location", "Item", "Amount"]
)

print("DATASET PREVIEW")
print(df.head())

#
df["Order_Time"] = pd.to_datetime(
    df["Order_Time"],
    format="%H:%M"
)
df["Hour"] = df["Order_Time"].dt.hour

peak_hours = df["Hour"].value_counts().sort_index()

# Popular Dishes
popular_items = df["Item"].value_counts()

# Revenue
revenue = df.groupby("Item")["Amount"].sum().sort_values(ascending=False)

# Dashboard
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

peak_hours.plot(kind="bar", ax=axes[0,0])
axes[0,0].set_title("Peak Order Hours")

popular_items.plot(kind="bar", ax=axes[0,1])
axes[0,1].set_title("Popular Dishes")

revenue.plot(kind="bar", ax=axes[1,0])
axes[1,0].set_title("Revenue by Dish")

df["Location"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    ax=axes[1,1]
)
axes[1,1].set_title("Orders by Location")

plt.tight_layout()
plt.show()

# Business Insights
print("\nBUSINESS INSIGHTS")
print("Peak Order Hour:", peak_hours.idxmax())
print("Most Popular Dish:", popular_items.idxmax())
print("Highest Revenue Dish:", revenue.idxmax())
print("Top Location:", df["Location"].value_counts().idxmax())
# Filtering Example

print("\nBIRYANI ORDERS")
biryani_orders = df[df["Item"] == "Biryani"]
print(biryani_orders.head())

print("\nCHENNAI ORDERS")
chennai_orders = df[df["Location"] == "Chennai"]
print(chennai_orders.head())
```
